```python
import matplotlib.pyplot as plt
#data
x = [1,2,3,4]
y = [10,20,25,30]
y2 = [30,25,20,15]
#Create figure and axes
fig, ax = plt.subplots()
#Plot data
ax.plot(x, y, label='Days', color='blue', linestyle='--')
ax.plot(x, y2, label='Temp', color='green', linestyle='-')
# Customise plot
ax.set_title('Sample Plot')
ax.set_xlabel('Days')
ax.set_ylabel('Temp')
ax.legend()
ax.grid(True)
ax.set_xlim(0, 5)
ax.set_ylim(0, 45)
ax.annotate('Peak', xy=(4, 30), xytext=(4, 30),
           arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=df['PetalLength'].head(100)
y=df['SepalLength'].head(100)
fig, (ax1,ax2) = plt.subplots(nrows=1,
                         ncols =2,
                         figsize=(5,6),
sharey=True)
ax1.set_title('iris data anylysis')
ax1.set_ylabel('Observed values')
ax1.violinplot(x)
ax2.set_title('Normal data anylysis')
ax2.violinplot(y)
plt.show()

```
